```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.polynomial import Polynomial


x = np.polynomial.Polynomial([2, 0.5, -6, -2, 2.5,1])
print(x)
print(x.roots()) # a

f = Polynomial([2, 0.5, -6, -2, 2.5,1])

def g3(x):
    m = Polynomial([-2, -0.5, 6, 2, 0, -1])
    return np.power(m(x)/2.5, 1.0/4.0)


a3 = 0.80
g3_a = []


for i in range(30):
    a3 = g3(a3)
    g3_a.append(a3)


xs = np.linspace(0, 2, 150)
ys = f(xs)

plt.axhline(y=0, color='r')
plt.plot(xs, f(xs), label='f(x)', color='y')
plt.plot(xs, g3(xs), label='g3(x)', color='m')


plt.legend()


if len(g3_a) > 0:
    root = np.array([g3_a[-1]])
    plt.plot(root, f(root), 'ko')

print(pd.DataFrame({'g3(x)': g3_a}))
```
